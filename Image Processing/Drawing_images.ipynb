**Import Libraries**

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

**Create Image Canvas**

```python
#creating empty image or canvas -> to be filled with pixel values
blank_img=np.zeros(shape=(512,512,3), dtype=np.int16) #512x512 pixels, 3-color channel
```

```python
plt.imshow(blank_img)
%matplotlib inline
```

**Draw Rectangle**

```python
#pt1: top left vertex, pt2: bottom right vertex, color -> RGB, thickness -> pixel
cv2.rectangle(blank_img,pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_img)
%matplotlib inline
#uneven pixels as was unable to fill-out all the pixels at the edge
```

**correcting edge problem**

```python
#pt1: top left vertex, pt2: bottom right vertex, color -> RGB, thickness -> pixel
cv2.rectangle(blank_img,pt1=(384,10), pt2=(480,150), color=(0,255,0), thickness=10)
plt.imshow(blank_img)
%matplotlib inline
```

**Square at Centre**

```python
cv2.rectangle(blank_img,pt1=(200,200), pt2=(300,300), color=(255,255,0), thickness=10)
plt.imshow(blank_img)
%matplotlib inline
```

**Draw Circles:**

```python
cv2.circle(img=blank_img, center=(100,100), radius=75, color=(0,255,0), thickness=10)
plt.imshow(blank_img)
```

**Filled-in Circle**

```python
cv2.circle(img=blank_img, center=(400,400), radius=75, color=(0,0,255), thickness=-1)
plt.imshow(blank_img)
#thickness: -1 for filled circle
```

**Draw lines**

```python
#pt1: top left, pt2: bottom right
cv2.line(blank_img, pt1=(0,0), pt2=(400,400),color=(102,255,200), thickness=5)
plt.imshow(blank_img)
```

**Writing Text on Image**

```python
#choose font
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
 #(image_write, text, org: bottom left corner)
cv2.putText(blank_img,text='Hello World', org=(10,500), fontFace=font,
            fontScale=2, color=(255,255,255), thickness =3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

**Custom Shape -Polygon**

```python
blank_img=np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(blank_img)
```

```python
#choose vertices
vertices=np.array([[100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
```

```python
# -1: place holder to evaluate # of rows (axis-0)
#  1: each row will contain 1-element (axis-1)
#  2: each array will contain 2-element (axis-2)
pts=vertices.reshape(-1,1,2)
pts
```

```python
# draw polygon with connected line- segment
cv2.polylines(blank_img,[pts], isClosed=True, color=(255,0,0), thickness=10)
plt.imshow(blank_img)
%matplotlib inline
```
