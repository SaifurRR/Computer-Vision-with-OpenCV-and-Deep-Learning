**Import Libraries**

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2 #open cv
%matplotlib inline
```

**Read & Display Image**

```python
#read image
img=cv2.imread('golden_retriever.jpg') #reads img directly as np array
type(img) # if path doesn't exist  -> type : NoneType
```

```python
#shape of the img
img.shape
```

```python
#display image
plt.imshow(img) # image in different color channel  -> mismatch between opencv & matplotlib
```

```python
# channel color code correction
# Matplotlib: RGB, openCV: BGR
matplot_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert openCV (color code) -> Matplotlib
plt.imshow(matplot_img)
```

```python
#read img in grayscale
img_gray=cv2.imread('golden_retriever.jpg', cv2.IMREAD_GRAYSCALE) #img_gray.min() -> 1 pure black, img_gray.max() -> 255 pure white;
img_gray.shape # no RGB dimension, dim: (height, width)
plt.imshow(img_gray, cmap='gray')  #convert openCV img color to gray
```

**Process Image**

```python
#resize image
res_img=cv2.resize(matplot_img, (1000,400)) #(src, destn_size : width, height)
plt.imshow(res_img) #dim opposite to openCV np.arr
```

```python
#resize by ratio
w_ratio=0.5
h_ratio=0.5
# reduce img by 50% of original
```

```python
res_img_ratio=cv2.resize(matplot_img,(0,0),matplot_img,w_ratio,h_ratio) # (original_img, tuple(0,0), original_img, width_ratio, height_ratio)
plt.imshow(res_img_ratio)
res_img_ratio.shape #dim reduced by 1/2
```

```python
#flip img : horizontal or vertical
flip_img_horizontal=cv2.flip(matplot_img,0) #horizontal flip
flip_img_vertical=cv2.flip(matplot_img,1) #vertical flip
flip_img_hor_ver=cv2.flip(matplot_img,-1) #horizontal & vertical flip
```

**Display & Save Image**

```python
#canvas for image:
fig=plt.figure(figsize=(18,10)) #fig_size in inches
ax1=fig.add_subplot(131)
ax1.title.set_text('horizontal flip')
ax1.imshow(flip_img_horizontal)

ax2=fig.add_subplot(132)
ax2.title.set_text('vertical flip')
ax2.imshow(flip_img_vertical)

ax3=fig.add_subplot(133)
ax3.title.set_text('horizontal & vertical flip')
ax3.imshow(flip_img_hor_ver)
```

```python
#save img file
cv2.imwrite('horizontal_vertical_flip.jpg', cv2.cvtColor(flip_img_hor_ver, cv2.COLOR_RGB2BGR)) # change color code from RGB 2 BGR
```
